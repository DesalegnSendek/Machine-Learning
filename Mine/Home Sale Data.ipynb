```python
# Example: Create a DataFrame
sample_data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 22, 28],
    "Gender": ["Female", "Male", "Male", "Male"],
    "Income": [60000, 75000, 50000, 80000]
}
df = pd.DataFrame(sample_data)

# Print the original DataFrame
print("Original DataFrame:")
print(df)

# --- CRUD Operations ---

# 1. Create (Insert) a New Row
new_row = {"Name": "Eva", "Age": 23, "Gender": "Female", "Income": 65000}
df = df.append(new_row, ignore_index=True)

# 2. Read (Retrieve) Data
# Example: Retrieve rows where Age is greater than 25
condition_age_gt_25 = df["Age"] > 25
filtered_df = df[condition_age_gt_25]

# 3. Update Data
# Example: Update the income for Bob
df.loc[df["Name"] == "Bob", "Income"] = 72000

# 4. Delete (Remove) Rows
# Example: Remove rows where Gender is "Male"
condition_gender_male = df["Gender"] == "Male"
df = df[~condition_gender_male]

# Print the modified DataFrame
print("\nModified DataFrame:")
print(df)
```
